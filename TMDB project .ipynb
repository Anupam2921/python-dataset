```python


df.info()
```

```python
# droping duplicates
df.drop_duplicates(inplace=True)
df.info()
```

```python
# keeping only those columns which are needed for further analysis
md = df[['popularity','budget','revenue','original_title','runtime','genres','vote_count','vote_average','release_year']]
```

```python
## creating profit variable
md['profit']=md['revenue']-md['budget']
```

```python
md.info()
```

```python
md.describe()
```

## to know about null value and calculate 

```python
md.isnull().sum()
```

```python
md.dropna(subset=['genres'],inplace=True)
# it will only remove the null values from the genres column,
```

```python
md.isnull().sum()
```

```python
md.head()
```

```python
# split genres and create a new entry for each of the genre a movie falls into
s = md['genres'].str.split('|').apply(pd.Series,1).stack()
s.index = s.index.droplevel(-1)
s.name = 'genres'
del md['genres']
md=md.join(s)
```

```python
md.head()
```

```python
md.shape
```

```python
corr =md.drop(['original_title','genres'],axis=1,inplace=False).corr()
sns.heatmap(corr,annot=True,cmap='BuPu')
```

## some research questions

## Q1; which genres are the most common (number of movies made)?

```python
num_movies_per_genre = md_genre.original_title.nunique().sort_values(ascending=False)
print(num_movies_per_genre)
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=num_movies_per_genre.values, labels= num_movies_per_genre.index,autopct='%1.1f%%')
plt.title('movies per genre in %')
plt.show()
```

## Q2. which genres have high avg. budget and revenue?

```python
sorted_by_budget= md_genre_budget.sort_values(by='budget',ascending=False)

md_genre_revenue=pd.DataFrame(md_genre['revenue'].mean()).reset_index()
sorted_by_revenue=md_genre_revenue.sort_values(by='revenue',ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x='genres',y="budget",data=sorted_by_budget)
plt.title("average movie budget by genre(highest first)")
plt.xlabel("genre")
plt.ylabel("average budget")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

## in the graphs above we closely see that the genre Adventure has both the highest avg. budget and revenue. fantasy come second in budget and revenue.

## Q3; which ganres have high avg.profit?

```python
sorted_by_profit=md_genre_profit.sort_values(by="profit",ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x="genres",y="profit",data=sorted_by_profit)
plt.title("average movie profit by genre(highest first)")
plt.xlabel("genre")
plt.ylabel("average profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## The top 5 genres in terms of avg.profit are:

### 1.Adventure
### 2. fantasy
### 3.Animation
### 4. family
### 5 science fiction

## Q4: which genre have high number of movies with an voting avg.>=8?


```python
md_vg_8= md[md['vote_average']>=8]
most_movies_vg_8=(pd.DataFrame(md_vg_8.groupby('genres').original_title.nunique())).sort_values('original_title',ascending=False).reset_index()
```

```python
most_movies_vg_8
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=most_movies_vg_8['original_title'],labels=most_movies_vg_8['genres'],autopct='%0.of%%')
plt.title('genres with voting average>=8')
plt.show()
```
